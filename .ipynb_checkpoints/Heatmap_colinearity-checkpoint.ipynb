Heatmap sur la colinearites des diff√©rentes colonnes : 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

merged_database = pd.read_csv("merged_database.csv")
# Calcul de la matrice de corr√©lation

# Exclure les colonnes non num√©riques
numeric_columns = merged_database.select_dtypes(include=['float64', 'int64']).columns
df_numeric = merged_database[numeric_columns]

# Calcul de la matrice de corr√©lation
correlation_matrix = df_numeric.corr()

# Cr√©ation de la heatmap avec seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Heatmap de Colinearit√© entre Colonnes')
plt.show()

```
